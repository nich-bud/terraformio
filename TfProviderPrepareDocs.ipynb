```julia
function TfProviderPrepareDocs(rootFolder::String)
    
    count = 0
    userNames = readdir(rootFolder)
    for userName in userNames

        repoNames = readdir(userName)
        for repoName in repoNames

            repoPath = joinpath(userName, repoName, "main")

            # Some folders are empty?
            if ! isdir(repoPath)
                continue
            end
            
            count += 1
            print(count)
            
            out = "tfplugindocs_stdout.txt"
            err = "tfplugindocs_stderr.txt"
            cd(repoPath)
            time = @elapsed run(pipeline(ignorestatus(`tfplugindocs`), stdout = out, stderr = err))
            cd("../../..")
            println("\t$time")
        end
    end
end
```
